```python
# リスト3.16 ヒストグラムを表示

# リスト3.16を実行のためBoston house-pricesの読み込みが必要
import keras
import matplotlib.pyplot as plt

(x_train, y_train), (x_test, y_test) = keras.datasets.boston_housing.load_data()


# matplotlibのグラフをインラインで表示する(ノート内でグラフを表示する)
%matplotlib inline

import matplotlib.pyplot as plt

# rcParams：matplotlibの設定値が入っているオブジェクト
plt.rcParams['font.size'] = 10*3 # 文字サイズ(デフォルト値：12)
plt.rcParams['figure.figsize'] = [18, 12] # グラフサイズ(デフォルト値：[8, 6])
plt.rcParams['font.family'] = ['IPAexGothic'] # フォント設定(デフォルト値：sans-serif)

plt.hist(y_train, bins=20)
plt.xlabel('住宅価格($1,000単位)')
plt.ylabel('データ数')
plt.show()
plt.plot(x_train[:, 5], y_train, 'o')
plt.xlabel('部屋数')
plt.ylabel('住宅価格($1000単位)')
```
